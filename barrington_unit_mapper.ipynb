```python
import pandas as pd
import numpy as np
```

```python
barrgrn = pd.read_csv("barrgrn.csv")
barrred = pd.read_csv("barrred.csv")
barrblue = pd.read_csv("barrblue.csv")
barrwhite = pd.read_csv("barrwhite.csv")
```

```python
barrgrn.head()
```

```python
barrred = barrred.rename(columns={"IO": "TYPE"})
barrred.head()
```

```python
barrblue.head()
```

```python
barrwhite.head()
```

```python
appended = barrgrn.append(barrred, ignore_index=True, 
                          sort=True).append(barrblue, ignore_index=True, 
                                            sort=True).append(barrwhite, ignore_index=True, sort=True)
appended
```

```python
mappings = pd.read_csv("uniques_barr.csv")
mappings.head()
```

```python
merged = appended.merge(mappings, how='left', left_on="UNITS", right_on="first")
merged = merged.dropna(axis=0, subset=['second'])
merged
```

```python
merged = merged.drop(columns=['UNITS', 'first'])
merged = merged.rename(columns={'second': 'UNITS'})
```
