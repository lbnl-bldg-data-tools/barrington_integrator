```python
#note that I had to encode the csv files in UTF-8 first, google if unsure how to do that
#PLEASE DO NOT RELABEL THE NAMES HERE! They need to be numbers right now, but will change later in the code.
barr = pd.read_csv("barrgrn.csv")
lsAuto = pd.read_csv("lstar77autoconfig.csv", names=np.arange(0, 10, 1))
#lsFinal is not something we need to use for merging purposes, rather serves as an ideal for what data should show up.
lsFinal = pd.read_csv("lstar77configfinal.csv", names=np.arange(0, 15, 1))
lsAuto
```

```python
unique_LS = barr.LS.unique()
table_LS_map = {}
for ls in unique_LS:
    tbl = barr[barr["LS"] == ls]
    table_LS_map[ls] = tbl
```

```python
#using the above line of code for possible LS numbers, calls all rows with respective LS number
LS_table = table_LS_map[77]
LS_table
```

```python
#cleans the table of // comments and of NaN values in column 0
def cleaning(ls_tbl):
    dropped_nan = ls_tbl.dropna(subset=[0])
    dropped_nan = dropped_nan[dropped_nan[0].str.contains("//")==False]
    return dropped_nan
cleaned = cleaning(lsAuto)
final_clean = cleaning(lsFinal)
final_clean
```

```python
#splits the cleaned table into their different tables
#depends on the fact that the names of the new tables have NaN values for all other columns other than col[0]
#RETURNS: a list of the different tables from the config files
def splitting(cleaned):
    reset = cleaned.reset_index(drop=True)
    #names = reset[reset[1].isna()==True][0].tolist()
    names_index = reset[reset[1].isna()==True][0].index
    tbls = []
    for i in range(len(names_index)):
        #tbl_name = reset.iloc[names_index[i]][0]
        col_names = reset.iloc[names_index[i] + 1].dropna().str.strip().tolist()
        if i != len(names_index)-1:
            tbl = reset.iloc[names_index[i]+2:names_index[i+1]].drop(np.arange(len(col_names), reset.iloc[0].size, 1), axis=1)
        else:
            tbl = reset.iloc[names_index[i]+2:].drop(np.arange(len(col_names), reset.iloc[0].size, 1), axis=1)
        tbl.columns = col_names
        tbls.append(tbl)
            
    return tbls
```

```python
#shows the names of the tables
reset = final_clean.reset_index(drop=True)
names = reset[reset[1].isna()==True][0].tolist()
names
```

```python
#labeling the tables for easier access, since lists get confusing
split = splitting(cleaned)
split
bridge = split[0]
data_arrays_1 = split[1]
data_arrays_2 = split[2]
connections = split[3]
adapters = split[4]
client_nodes = split[5]
server_nodes = split[6]
client_map = split[7]
server_map = split[8]
```

```python
#deleting CI types
merge_test = LS_table.copy()
merge_test = merge_test[merge_test['TYPE'] != "CI"]
merge_test['MS'] = merge_test['MS'].astype(int)
#mapping barr types to config types
merge_test["TYPE"] = merge_test["TYPE"].map({"DI": "Bi", "DO": "Bo", "SP": "Av", "AI": "Ai", "AO": "Ao"})
#now changing autoconfig map_descriptor_name to MS number, also removing NodeOnline values
right_merge = server_map.copy()
#removing random white spaces from importing
right_merge = right_merge.apply(lambda x: x.str.strip())
right_merge = right_merge[right_merge["Map_Descriptor_Name"] != 'NodeOnline']
right_merge["Map_Descriptor_Name"] = right_merge["Map_Descriptor_Name"].str.extract(r'(\d{2})', expand=False).astype(int)
right_merge["Data_Array_Offset"] = right_merge["Data_Array_Offset"].astype(int)
```
